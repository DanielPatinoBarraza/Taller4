# 1 Analisis de datos

Este archivo sera un vistazo al comportamiento del SyP500 en el años 2019 en contrasnte del 2020

```python
#1.1 Carga de datos
Data= pd.read_csv("SyP.csv", parse_dates=["Date"]) #Esta sera 2019
Data1=pd.read_csv("C:/Users/AlejoP/Downloads/SyP.csv", parse_dates=["Date"]) #Esta sera 2019
#Los importamos y los descarga
```

```python
#1.2 Inspeccion de la informacion
#Data.info()
#print(Data)
#Data.head()
```

```python
#1.3 y 1.4 Limpieza de datos y Modificacion de datos, estas funciones solo se pueden ejecutar una ves ya que genera un error al borrar algo que ya esta borrado
Data.drop(["Adj Close", "Open", "Volume","High","Low"],  axis='columns', inplace=True) #Para el analisis de la caida del precio del SyP 500 realmente solo se necesita el Precio de cierre ajustado
Data1.drop(["Adj Close", "Open", "Volume","High","Low"],  axis='columns', inplace=True)
Data.drop(range(252, 344,1), axis=0,inplace=True)
Data1.drop(range(0, 252,1), axis=0,inplace=True) #Esto es para separar los datos del 2019 de los de 2020, el año financiero tiene 252 dias

Data.set_index("Date",inplace=True) #Para mi es mas util que mi indice
Data1.set_index("Date",inplace=True) 
#Data1.set_index("Date",inplace=True)
```

```python
#1.5 Modelacion
plt.figure(figsize=(20,5))
plt.plot(Data.index,Data.Close,"r.-" ,label="Close Prices")
plt.title("Indice SyP500 2019",fontdict={"fontweight":"bold","fontsize":"15"})
plt.legend()
plt.xticks(Data.index[::20])
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show() 
#plt.savefig("SyP500_2019.jpg",dpi=300)
plt.figure(figsize=(20,5))
plt.title("Indice SyP500 2020",fontdict={"fontweight":"bold","fontsize":"15"})
plt.plot(Data1.index,Data1.Close,"b.-", label="Close Prices")
plt.xticks(Data1.index[::10])
plt.legend()
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
plt.style.use("default")
labels= ["2019","2020"]
plt.boxplot([Data.Close,Data1.Close], labels=labels)
plt.title("Comparacion SyP500 2019 - 2020")
plt.ylabel("Close Prices")
plt.show()
```

1 Inicial mente se puede notar como la caida del mercado en el 2020 fue mayor que todos las utlidades del 2019 fue una caida que practicamente podria atrasar el mercado un año ya que el mercado actualmente tiene el mismo nivel que tenia en mayo del 2019

2 al ver la caja de bigotes nos damos cuenta que realmente la caida no a sido tan impactante ya que el promedio del precio sigue siendo superior al promedio del año pasado y que mas del 50% de los datos son mayores a ese promedio
